# Dataset

Iris plants dataset is a benchmark dataset for testing supervised machine learning problems. It comprises of four attributes (sepal length, sepal width, petal length, petal width) and 50 samples from each of the three different species (classes) of the iris flower. 








```
#Import libraries

from sklearn.datasets import load_iris
from sklearn import preprocessing
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

  *   **For Exploration of Attributes and label names I used inbuilt function of Sklearn dataset.**

```
# Load the iris dataset together with the target labels
iris = load_iris()
print(iris.feature_names) #shows coulmns(ie attributes) Heading
print(iris.data) #Attribute Values
print(iris.target_names) #Label Names
print(iris.target) #Label Values
```


Further explore the dataset by finding the minimum, maximum, mean and standard deviation for each attribute.  **(4 points)** {1*4=4}
*   **Here to calculate min, max, std of each attribute, tranformed dataset into Pandas Dataframe "iris_d". after that called pandas funtion for calculating min,max and std value of each attribute. Created function "exp" (exp stands for exploration) for finding value of different values of a attribute.**

```
 #Look at first 10 elements of the data
print(iris.data[:10], "\n")

#Compute min, max, mean and std for each attribute {for all 150 samples}

def exp(iris_d, attrib): #Function to print min,max, mean, std of a attrib
  print("Minimum of ", attrib, " ", iris_d[attrib].max())
  print("Maximum of ", attrib, " ", iris_d[attrib].min())
  print("Mean of ", attrib, " ", iris_d[attrib].mean())
  print("Standard Deviation of ", attrib, " ", iris_d[attrib].std(), "\n")

iris_d = pd.DataFrame(data = iris.data, columns=iris.feature_names) #iris_d is pandas dataframe with values of iris dataset
# iris_d["target"] = iris.target
for i in list(iris_d):
  exp(iris_d, i)
```

Data Normalization rescales the attribute values to the range of [0,1]
1. Normalize  (min-max normalization) the data and print the normalized data in a tabular form.  **(1 point)** 
2. Plot the histogram corresponding to each attribute as done above. **(4 points)** {1*4=4}

```
# Normalize the dataset 

#Using MinMaxScaler from sklearn.preprocessing for (min-max) Normalization of the given dataset.
min_max_scaler = preprocessing.MinMaxScaler()
iris_norm = min_max_scaler.fit_transform(iris_d)
iris_norm = pd.DataFrame(iris_norm, columns=iris.feature_names)
# Print dataset before normalization    
print("Dataset Before Normalization\n")      
display(iris_d.style) #Non Normalized Dataset

# Print dataset after normalization  
print("\n\n\n Dataset After Normalization")
display(iris_norm.style) #Normalized Dataset

print("\n\n")

# For entire normalized dataset, plot 4 histograms for sepal length (cm), sepal width (cm), petal length (cm) and petal width (cm)
for i in list(iris_norm):
  iris_norm.hist(column = i)  #Plotting Histogram of Different attributes of Normalized dataset using Pandas Library.
```
